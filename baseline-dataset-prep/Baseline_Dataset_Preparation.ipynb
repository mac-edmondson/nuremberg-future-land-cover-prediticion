```python
import pandas as pd
```

```python
# Load the dataset ESA World Cover 2020 or 2021
df_2020 = pd.read_csv('/content/Nuremberg_complete_2020_dataset_with_all_labels(1).csv')
```

```python
# Drop Columns that is unnesscary
df_2020.drop(columns=['system:index','image_count'], inplace=True)
```

```python
# Fill missing values with mean
df_2020.fillna(df_2020.mean(numeric_only=True), inplace=True)
```

```python
# Converted the Dtypes of count column from float64 to int64
cols = [ 'B11_count','B12_count','B1_count','B2_count','B3_count','B4_count','B5_count','B6_count','B7_count','B8A_count','B8_count','B9_count',]
df_2020[cols] = df_2020[cols].astype(int)
```

```python
# Load the dataset ESA World Cover 2016,2017,2018,2019
df_2019 = pd.read_csv('/content/data_2019.csv')
```

```python
# Drop Columns that is unnesscary.**If required**, Adjust the .drop(columns=))
# Converted the Dtypes of count column from float64 to int64
# Fill missing values with mean
df_2019.drop(columns=['system:index','image_count','Unnamed: 111','Unnamed: 112'], inplace=True)
df_2019.fillna(df_2019.mean(numeric_only=True), inplace=True)
df_2019[cols] = df_2019[cols].astype(int)
```

```python
# Generating the Combined dataset joining on .geo column
df_combined = df_2019.merge(df_2020[['bare_sparse_vegetation',
 'built_up',
 'cropland',
 'grassland',
 'herbaceous_wetland',
 'mangroves',
 'moss_lichen',
 'shrubland',
 'snow_ice',
 'tree_cover',
 'water',
 '.geo']],on='.geo', how='inner')
```

```python
# Verify the Shapes of your columns
print(df_2020.shape)
print(df_2019.shape)
print(df_combined.shape)
```

```python
# Generate the CSV File
df_combined.to_csv('df_2019_mit_2020_labels.csv', index=False)
```
